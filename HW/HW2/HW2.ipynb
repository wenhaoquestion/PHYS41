```python
def mysum(N):
    if N==1:
        return 1
    else:
        return N+mysum(N-1)
```

```python
def sort(items):
    nums = [x for x in items if isinstance(x, (int, float))]
    strs = [x for x in items if isinstance(x, str)]
    return sorted(nums) + sorted(strs)
```

```python
test = [78, 'question', 6.9, 'mac', 114514, 'nvida', -1]
sort(test)
```

```python
def calc(lst):
    result = lst[0]
    for i in range(1, len(lst), 2):
        if lst[i] == '+':
            result += lst[i+1]
        elif lst[i] == '-':
            result -= lst[i+1]
        elif lst[i] == '*':
            result *= lst[i+1]
        elif lst[i] == '/':
            result /= lst[i+1]
    return result
```

```python
list1c = [2, '+', 4, '-', 5, '*', 10]
calc(list1c)
```

```python
def calc2(lst):
    t = [lst[0]]
    i = 1
    while i < len(lst):
        op = lst[i]
        num = lst[i+1]
        if op in ('*', '/'):
            left = t.pop()
            t.append(left * num if op == '*' else left / num)
        else:
            t.append(op)
            t.append(num)
        i += 2

    value = t[0]
    i = 1
    while i < len(t):
        op = t[i]
        num = t[i+1]
        if op == '+':
            value += num
        elif op == '-':
            value -= num
        i += 2

    return value
```

```python
list1d = [2, '+', 4, '-', 5, '*', 10]
calc2(list1d)
```

**(b)**

```python
def solve_quad(a, b, c):
    a = np.asarray(a)
    b = np.asarray(b)
    c = np.asarray(c)

    disc = b*b - 4*a*c
    disc = disc.astype(np.complex128)
    sqrt_disc = np.sqrt(disc)

    x1 = (-b + sqrt_disc) / (2*a)
    x2 = (-b - sqrt_disc) / (2*a)
    return x1, x2
```

```python
print(solve_quad(1, 0, 1),"\n")
print(np.array([11,1,1]), np.array([6,7,2]), np.array([1,1,1]))
```
