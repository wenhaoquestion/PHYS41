### Problem 2: Basic Functions

**(a)**

```python
def solution_a(lst, item):
    count = 0
    for x in lst:
        if x == item:
            count += 1
    return count
```

```python
list_a=['cat','dog',1,2,'bear',2,4,1]
print(solution_a(list_a,'cat'))
print(solution_a(list_a,2))
print(solution_a(list_a,3))
```

**(b)**

```python
def solution_b(nums):
    total = 0
    product = 1
    for x in nums:
        total += x
        product *= x
    return (total, product)
```

```python
list_b=[1.1,2.1,3.1,4.4,1.4,1.9,100,12.]
solution_b(list_b)
```

**(c)**

```python
def solution_c(lst):
    total = 0
    for x in lst:
        if type(x) == int or type(x) == float:
            total += x
        else:
            print("Error: not a float or int:", x, "type:", type(x))
    return total
```

```python
list_c = [1, 2, 3, 'a', 4, 5, 'b', 6]
solution_c(list_c)
```

**(d)**

```python
def solution_d(my_class, student_name, class_name):
    if student_name not in my_class:
        return "Error: there is no this student in this class"
    if class_name not in my_class[student_name]:
        return "Error: class not found for this student"
    return my_class[student_name][class_name]
```

```python
my_class={'Alice': {'Phys4':'A','Phys41':'A+'}}
my_class['Alice']
```

```python
print(solution_d(my_class, 'Alice', 'Phys41'))
print(solution_d(my_class, 'Andy', 'Phys41'))
print(solution_d(my_class, 'Alice', 'Math20D'))
```
