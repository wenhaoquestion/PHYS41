### **Question 1: Setup and Weight Initialization**

```python
import numpy as np
np.random.seed(42)
X = np.random.randn(128, 16)
print(X.shape)
```

```python
W_Q = np.random.randn(16, 16)
W_K = np.random.randn(16, 16)
W_V = np.random.randn(16, 16)

print(W_Q.shape)
print(W_K.shape)
print(W_V.shape)
```

### **Question 2: Linear Projections (Creating Q, K, V)**

```python
Q = X @ W_Q
K = X @ W_K
V = X @ W_V

print(Q.shape)
print(K.shape)
print(V.shape)
```

### **Question 3: The Attention Scores**

```python
d_k = 16

scores = Q @ K.T
print(scores.shape)
```

### **Question 4: The Softmax (Probability Distribution)**

```python
def softmax(x):
    exp_x = np.exp(x)
    sum_exp = np.sum(exp_x, axis=1, keepdims=True)
    return exp_x / sum_exp
```

```python
attention_weights = softmax(scores)
print(attention_weights.shape)
print(np.sum(attention_weights, axis=1))
```

### **Question 5: The Final Representation**

```python
output = attention_weights @ V

print(output.shape)
```
