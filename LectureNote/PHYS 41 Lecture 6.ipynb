## np.where
find the index of elements in array that satisfies certain condition

```python
in1 = np.array([10,4,-2])
np.where(in1>0)
```

You can use an numpy array of indices to index an numpy array
- [0,2,3] is a python list
- np.array([0,2,3]) turns this into a numpy array
- We can use this new numpy array to index our existing numpy array: a[np.array([0,2,3])]

```python
print(a)
print(a[np.array([0,2,3])])
```

The good thing is that np.where also provide you an numpy array of indices (that satisfy certain conditions), this means we can use np.where to index numpy array as well

```python
print(in1[np.where(in1>0)])
```

But this is basically what we did with the boolean array!

```python
flag = in1>0
print(flag)
print(in1[flag])
print(in1[in1>0]) # This line should give me exactly the same thing as the previous cell
```

```python
print(in1[np.where(in1>0)]) # for a more direct comparison
print(in1[in1>0])
```

Why do we need np.where? Because it can take additional argument that is very useful.
- np.where(condition, out_true, out_false)
- out_true is what you do when the condition returns true
- out_false is what you do when the condition returns false

```python
print(in1)
print(np.where(in1>0,5,-100 ))
```

**Revisit this problem**: I want to output an array so that when a<b I have a\**2 and when a>= b I have b\**2

## Additional Numpy Logics Operations
- we just discussed `np.logical_not`
- There are other np.logical operations, for example `np.logical_and` and `np.logical_or`
- `np.logical_or` compare two boolean np array (to generate boolean np array one way is to compare two np arrays)

```python
test1 = np.array([0,5,-10])
test2 = np.array([-1,1,-5])
print(test1<test2)
print(test2>0)
print("np.logical_or",np.logical_or(test1<test2, test2>0))
print("np.logical_and",np.logical_and(test1<test2, test2>0))
```

## Using the "Try" command
there will be many times you try to do something but you don't know if it will works. Instead of spitting out an error message, you can simply as python to "ignore" the error.

```python
int("hi")
```

```python
try:
    int("hi")
except:
    print("oops, something goes wrong")
```

```python
data = ["1", "2", "hi", "3"]
```

```python
def list_to_int(listin):
    out = listin.copy()
    for i in range(len(listin)):
        try:
            int(listin[i])
        except:
            print(f"Convert to integer failed! {listin[i]} cannot be converted into integer")
        else:
            out[i] = int(listin[i])
    return out
```

```python
list_to_int(data)
```

This is a very useful functionalities when you work with mixed data types

```python
def keep_ints(listin):
    out = []
    for i in range(len(listin)):
        try:
            out.append(int(listin[i]))
        except:
            print(f"Convert to integer failed! {listin[i]} cannot be converted into integer")
    return out
```

```python
keep_ints(data)
```

You can also append error typ to the except command so that it only captures certay type of error.

```python
def test_error_capture(listin):
    out = []
    for i in range(len(listin)):
        try:
            out.append(int(listin[i]))
        except EOFError:
            print(f"Convert to integer failed! {listin[i]} cannot be converted into integer")
    return out
```

```python
test_error_capture(data)
```
