## Variable and Variable Assignment
- Variable assigment can be achieven with a = sign
- Math, 1+1=2 and 2=1+1 are the same equation, which means the = sign itself is non-directional
- In python and other programming languages, the = sign has a different meaning and at the same time it is "directional": **the direction is right to left, and it assigns the value at the right of the equal sign to the left**

```python
x = 1
```

```python
print(x)
```

```python
print(y)
```

```python
x = x + 1
print(x)
```

this expression can be further simplified as:

```python
print(x)
x += 1
print(x)
# x += 1 is exactly the same as x = x + 1
```
