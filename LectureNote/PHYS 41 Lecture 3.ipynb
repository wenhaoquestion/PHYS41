## Loop through a dictionary

```python
dict1_keys = dict1.keys()
print(dict1_keys)
```

```python
def factorial(n):
    out = 1
    if n > 0: # we need to know if the input is a positive integer for factorial to be well-defined
        for i in range(int(n)):
            out *= (i+1)
    return out
```

Return statement is useful because it immediately terminate the function

```python
print(factorial(5), factorial(7), factorial(6.6))
```

720 is actually 6!, so basically the .6 part of 6.6 is ignored. This make sense because in my factorial function I convert any floating point number into integers

To calculate the factorial of fractional number, we need to use gamma function.

Gamma[x] = (x-1)Gamma[x-1]

and 

Gamma[1] = 1

So we know that Gamma[n] = (n-1)!

```python
def Gamma(n):
    if n <= 1:
        return 1
    else:
        return (n-1)*Gamma(n-1) #This is recursion
```

**Recursion** means a function calls itself

```python
print(Gamma(5+1), Gamma(3.2))
```

## Sorting algorithm

```python
def dumb_sort(alist):

    b = [] # I'm making a place to put my numbers in order, this is the out list

    c = alist[:] # I'm making a copy of my original list

    while len(c)>0: #I'm going to keep removing numbers from my c list and put them into the b list until c is empty

        largest = c[0]

        for item in c: # run over all elements in the c list
            if (item > largest): # if the current element (item) is larger than the stored largest value
                largest = item # replace largest with the current element
        b.append(largest) # adding the largest values figured out from the c list
        c.pop(c.index(largest)) # removing the largest values from the c list, this is important because our while loop is based on c list size
    return b
```

```python
unordered = [3,5,1,5,2,4,5,2,1]
```

```python
dumb_sort(unordered)
```

```python
import time
tstart = time.time()
dumb_sort(unordered*10000)
print(time.time() - tstart)
```
