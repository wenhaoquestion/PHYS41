### Loop through the list
There are actually 2 ways to loop through the list, both of them use a for loop:
- first way:


```python
for item in list_example:
    print(item)
```

- second way:

```python
for index in range(len(list_example)):
    print(list_example[index])
```

- the _len()_ function just tells you the length of your list
- the _range()_ function provides me a list of indcies based on the length of the list
   - range takes 3 argument: range(start = 0, stop, increment = 1)
   - If you call range(8), that means range(start=0, stop=8, increment=1)
   - **important**: range actually does not include the stop argument we give it, for example if you call range(8), it only iterate through [0, 1, 2, 3, 4, 5, 6, 7]

```python
print(len(list_example))
```

```python
print(list(range(8)))
```

I want to try a few different range operations

```python
print(list(range(2,8)))
```

**important**: the start argument of range is included in the list/iterator

```python
print(list(range(2,8,2)))
```

```python
for index in range(2, len(list_example), 2):
    print(list_example[index])
```

### Nested For Loop

```python
list2 = [1,"b", 3]
for item in list2: # outer for loop
    for item2 in list2: # inner for loop
        print(f"item1 is {item} and item2 is {item2}")
    print(f"item1: {item}")
print("==========all done========")
```

The first step in outer for loop provides me:

item1 is 1 and item2 is 1

item1 is 1 and item2 is b

item1 is 1 and item2 is 3

item1: 1

The second step in outer for loop provides me:

item1 is b and item2 is 1

item1 is b and item2 is b

item1 is b and item2 is 3

item1: b

Thew 3rd step in outer for loop provides me

item1 is 3 and item2 is 1

item1 is 3 and item2 is b

item1 is 3 and item2 is 3

item1: 3

**The way we define for loop and their scope is based on indentation**

### Slicing a list
- indexing means we only want to get 1 element ouf of my list
- Slicing means I want to get a subsets of elements (in certain order) from my list

```python
list_example[:5] #list_example[0:5:1]
```

List slicing is achieven with the semicolon notation **::**
- _start:stop:increment_
- similar to the range() logic, start is by default 0 and increment is by default 1, stop is by default the length of the list
- the second semicolon can be omitted if we do not plan to change the increment

```python
print(list_example[:5])
print(list_example[0:5:1])
```

```python
print(list_example[2:]) #identical to list_example[2:last:1]
```

what if I want to slice my list from the start to the end but with non-1 increment

```python
print(list_example[::2]) # identical to list_example[0:last:2]
```

You can also use this functionality to loop back through the list

```python
print(list_example[::-1])
```

```python
print(list_example[7:2:-1])
```

```python
list_example[-1]
```

the very last element in your list is indexed as -1
the second last element in your list is indexed as -2
.....

```python
print(list_example[:-1])
print(list_example[:-2])
```
