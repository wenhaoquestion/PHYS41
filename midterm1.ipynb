```python
def mysort(input_list):
    arr = input_list[:]
    n = len(arr)

    for i in range(n):
        for j in range(i+1, n):
            if arr[j] > arr[i]:
                arr[i], arr[j] = arr[j], arr[i]

    return arr
```

```python
def mysort2(input_list):
    arr = input_list[:]
    n = len(arr)

    for i in range(n):
        max_idx = i
        for j in range(i + 1, n):
            if arr[j] > arr[max_idx]:
                max_idx = j
        arr[i], arr[max_idx] = arr[max_idx], arr[i]

    return arr
```

```python
print(mysort([10.2,11.4,5.14]))
print(mysort2([10.2,11.4,5.14]))
```

```python
def sort_names_by_height(nba_data):
    heights = []
    for name in nba_data:
        heights.append(nba_data[name])

    sorted_heights = mysort(heights)

    result = []
    for h in sorted_heights:
        for name in nba_data:
            if nba_data[name] == h:
                result.append(name)
                break

    return result
```
